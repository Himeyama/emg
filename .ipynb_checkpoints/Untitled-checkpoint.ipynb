```python
from exe import bandpass
import numpy as np
import matplotlib.pyplot as plt

f = 1925.926

ys = np.loadtxt("tmp.csv", delimiter=",").T
filtered = np.zeros((ys.shape))
x = np.arange(len(ys[0])) / f
for y in ys:
    y2 = bandpass.bandpass(y, f, np.array([10, 15]), np.array([3, 25]), 2, 30)
    y2abs = np.abs(y2)
    plt.plot(x, y2abs)
    plt.show()
```
